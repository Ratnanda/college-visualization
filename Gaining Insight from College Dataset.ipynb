# Gaining Insight from College Dataset

Data ini didapatkan dari website kaggle.com

```python
df = pd.read_csv ("D:/Code/Data/College.csv")
df.head()
```

```python
#deteksi missing value
np.sum(df.isnull())
```

```python
#fill missing value with mean
df['Economic Need Index']=df['Economic Need Index'].fillna(df['Economic Need Index'].mean())
df['School Income Estimate']=df['School Income Estimate'].fillna(df['School Income Estimate'].mean())
df['Average ELA Proficiency']=df['Average ELA Proficiency'].fillna(df['Average ELA Proficiency'].mean())
df['Average Math Proficiency']=df['Average Math Proficiency'].fillna(df['Average Math Proficiency'].mean())
np.sum(df.isnull())
```

```python
df['Rigorous Instruction Rating']=df['Rigorous Instruction Rating'].fillna(df['Rigorous Instruction Rating'].mode()[0])
df['Collaborative Teachers Rating']=df['Collaborative Teachers Rating'].fillna(df['Collaborative Teachers Rating'].mode()[0])
df['Supportive Environment Rating']=df['Supportive Environment Rating'].fillna(df['Supportive Environment Rating'].mode()[0])
df['Effective School Leadership Rating']=df['Effective School Leadership Rating'].fillna(df['Effective School Leadership Rating'].mode()[0])
df['Strong Family-Community Ties Rating']=df['Strong Family-Community Ties Rating'].fillna(df['Strong Family-Community Ties Rating'].mode()[0])
df['Trust Rating']=df['Trust Rating'].fillna(ets['Trust Rating'].mode()[0])
df['Student Achievement Rating']=df['Student Achievement Rating'].fillna(df['Student Achievement Rating'].mode()[0])
np.sum(df.isnull())
```
